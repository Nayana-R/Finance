```python
def toFloat(str):
    try:
        x = float(str)
    except:
        x = None
    return x
```

# Future Value, Present Value

```python
def Value_without_Annuity():
    print('FUTURE VALUE or PRESENT VALUE WITHOUT ANNUITY')
    print('What is to be Calculated :')
    print('1. Present Value\n2. Future Value\n3. Rate\n4. Time Period')
    calc = int(input('Enter Choice : '))
    def findFV():
        print('Future Value without Annuity')
        pv = toFloat(input('Present Value : '))
        r = toFloat(input('Rate as (Rate/100) : '))
        n = toFloat(input('Time period : '))
        t = toFloat(input('No of Compounding Period : '))
        fv = round(pv * (1+(r/t)) ** (n*t), 4)
        print('Future Value : ', fv)
        print('\n')

    def findPV():
        print('Present Value without Annuity')
        fv = toFloat(input('Future Value : '))
        r = toFloat(input('Rate as (Rate/100) : '))
        n = toFloat(input('Time period : '))
        t = toFloat(input('No of Compounding Period : '))
        pv = round(fv / ((1 + (r/t)) ** (n*t)), 4)
        print('Present Value : ', pv)
        print('\n')

    def findN():
        print('Time Period')
        pv = toFloat(input('Present Value : '))
        fv = toFloat(input('Future Value : '))
        r = toFloat(input('Rate as (Rate/100) : '))
        t = toFloat(input('No of Compounding Period : '))
        n = round((1/t) * (log(fv / pv) / (log(1 + (r/t)))), 2)
        print('Time period : {0} years' .format(n))
        print('\n')
        
    def findR():
        print('Rate of Interest')
        pv = toFloat(input('Present Value : '))
        fv = toFloat(input('Future Value : '))
        n = toFloat(input('Time period : '))
        t = toFloat(input('No of Compounding Period : '))
        r = round(t * (((fv / pv) ** (1.0/(n*t))) - 1), 2)
        print('Rate : {0} %' .format(r*100))
        print('\n')

    val = None 

    if calc is 1:
        val = findPV()
    elif calc is 2: 
        val = findFV()
    elif calc is 3:
        val = findR()
    elif calc is 4:
        val = findN()
    else:
        val = 'Invalid Choice'

    return val
```

# Present Value Annuity

```python
def PVAnnuity():
    print('PRESENT VALUE WITH ANNUITY')
    print('A. End of the Year (Ordinary)\nB. Beginning of the Year (Annuity Due)')
    choice = input('Enter Choice : ')
    if choice == 'A':
        print('Present Value with Annuity at the End of the Year')
        print('What is to be Calculated :')
        print('1. Present Value with Annuity\n2. Annuity\n3. Time Period')
        calc = int(input('Enter Choice : '))
        def findPVA():
            a = toFloat(input('Annuity : '))
            r = toFloat(input('Rate as (Rate/100) : '))
            n = toFloat(input('Time period : '))
            t = toFloat(input('No of Compounding Period : '))
            pva = round(a / r * (1 - (1 + (r/t)) ** (-n*t)), 4)
            print('Present Value with Annuity (Ordinary) : ', pva)
            print('\n')

        def findA():
            r = toFloat(input('Rate as (Rate/100): '))
            n = toFloat(input('Time period : '))
            t = toFloat(input('No of Compounding Period : '))
            pva = toFloat(input('Present Value with Annuity : '))
            a = round((pva * (r/t)) / (1 - (1 + (r/t)) ** (-n*t)), 4)
            print('Annuity : ', a)
            print('\n')
            
        def findN():
            a = toFloat(input('Annuity : '))
            r = toFloat(input('Rate as (Rate/100) : '))
            t = toFloat(input('No of Compounding Period : '))
            pva = toFloat(input('Present Value with Annuity : '))
            nt = round(-1 * log(1 - (pva * (r/t) / a)) / log(1 + (r/t)), 2)
            n = round((-1 / t) * log(1 - (pva * (r/t) / a)) / log(1 + (r/t)), 2)
            print('No of Compounding Period(s) : ', nt)
            print('Time period : {0} years' .format(n))
            print('\n')
            
        val = None 

        if calc is 1:
            val = findPVA()
        elif calc is 2:
            val = findA()
        elif calc is 3:
            val = findN()
        else:
            val = 'Invalid Choice'
        
        return val
            
    elif choice == 'B':
        print('Present Value with Annuity at the Beginning of the Year')
        print('What is to be Calculated :')
        print('1. Present Value with Annuity\n2. Annuity\n3. Time Period')
        calc = int(input('Enter Choice : '))
        def findPVA():
            a = toFloat(input('Annuity : '))
            r = toFloat(input('Rate as (Rate/100) : '))
            n = toFloat(input('Time period : '))
            t = toFloat(input('No of Compounding Period : '))
            pva = round((a / (r/t) * (1 + (r/t))) * (1 - (1+(r/t)) ** (-n*t)), 4)
            print('Present Value with Annuity (Due) : ', pva)
            print('\n')
            
        def findA():
            r = toFloat(input('Rate as (Rate/100) : '))
            n = toFloat(input('Time period : '))
            t = toFloat(input('No of Compounding Period : '))
            pva = toFloat(input('Present Value with Annuity : '))
            a = round((pva * (r/t)) / ((1 - (1 + (r/t)) ** (-n*t)) * (1 + (r/t))), 4)
            print('Annuity : ', a)
            print('\n')

        def findN():
            a = toFloat(input('Annuity : '))
            r = toFloat(input('Rate as (Rate/100) : '))
            t = toFloat(input('No of Compounding Period : '))
            pva = toFloat(input('Future Value with Annuity : '))
            nt = round(-1 * log(1 - ((pva * (r/t)) / (a * (1 + (r/t))))) / log(1 + (r/t)), 2)
            n = round((-1 / t) * log(1 - ((pva * (r/t)) / (a * (1 + (r/t))))) / log(1 + (r/t)), 2)
            print('No of Compounding Period(s) : ', nt)
            print('Time period : {0} years' .format(n))
            print('\n')
        
        val = None 

        if calc is 1:
            val = findPVA()
        elif calc is 2:
            val = findA()
        elif calc is 3:
            val = findN()
        else:
            val = 'Invalid Choice'

        return val
```

# Perpectuity

```python
def Perpectuity():
    print('PERPECTUITY')
    print('What is to be Calculated :')
    print('1. Present Value\n2. Annuity\n3. Rate')
    calc = int(input('Enter Choice : '))
    def findPV():
        a = toFloat(input('Annuity : '))
        r = toFloat(input('Rate as (Rate/100) : '))
        pv = round(a / r, 4) 
        print('Present Value : ', pv)
        print('\n')

    def findA():
        r = toFloat(input('Rate as (Rate/100) : '))
        pva = toFloat(input('Present Value Annuity : '))
        a = round(pva * r, 4)
        print('Annuity : ', a)
        print('\n')
                
    def findR():
        a = toFloat(input('Annuity : '))
        pva = toFloat(input('Present Value Annuity : '))
        r = round(a / pva, 4)
        print('Rate : {0} %' .format(r))
        print('\n')

    val = None 

    if calc is 1:
        val = findPV()
    elif calc is 2:
        val = findA()
    elif calc is 3:
        val = findR()
    else:
        val = 'Invalid Choice'

    return val
```

# Net Present Value

```python
def NPV():
    print('NET PRESENT VALUE')
    print('What is to be Calculated :')
    print('1. Net Present Value\n2. Initial Investment\n3. Sum of Present Values (of future cash flows)')
    calc = int(input('Enter Choice : '))
    def findNPV():
        ls = []
        sum_pv = 0
        ini = toFloat(input('Inital investment : '))
        n = int(input('Time Period : '))
        r = toFloat(input('Rate as (Rate/100) : '))
        
        print('Enter year wise Cash Inflow : ')
        
        for yr in range(1, n+1):
            amount = toFloat(input())
            ls.append(amount)
            sum_pv = sum_pv + amount / (( 1 + r ) ** yr)
            
        npv = round(sum_pv - ini, 4)
        print('Net Present Value : ', npv)
        print('\n')
        
    def findINI():
        ls = []
        sum_pv = 0
        n = int(input('Time Period : '))
        r = toFloat(input('Rate as (Rate/100) : '))
        npv = toFloat(input('Net Present Value : '))
        
        print('Enter year wise Cash Inflow : ')
        
        for yr in range(1, n+1):
            amount = toFloat(input())
            ls.append(amount)
            sum_pv = sum_pv + amount / (( 1 + r ) ** yr)
            
        ini = round(sum_pv - npv, 4)
        print('Initial Amount : ', ini)
        print('\n')
        
    def findSum_PV():
        ls = []
        sum_pv = 0
        n = int(input('Time Period : '))
        r = toFloat(input('Rate as (Rate/100) : '))
        
        print('Enter year wise Cash Inflow : ')
        
        for yr in range(1, n+1):
            amount = toFloat(input())
            ls.append(amount)
            sum_pv = sum_pv + amount / (( 1 + r ) ** yr)
            
        print('Sum of Present Values of future Cash Flows : ', sum_pv)
        print('\n')
           
    val = None
    
    if calc is 1:
        val = findNPV()
    elif calc is 2:
        val = findINI()
    elif calc is 3:
        val = findSum_PV()
    else:
        val = 'Invalid Choice'

    return val
```

# Profitability Index

```python
def PI() :
    print('PROFITABILITY INDEX')
    ls = []
    n = int(input('Time Period : '))
    r = toFloat(input('Rate as (Rate/100) : '))
    sum_pv = 0
    ini = toFloat(input('Inital investment : '))
    
    print('Enter year wise Cash Inflow : ')
                      
    for yr in range(1,n+1):
        amount = float(input())
        ls.append(amount)
        sum_pv = sum_pv + amount / (( 1 + r ) ** yr)
        
    pi = round(sum_pv / ini, 4)
    print('Profitability Index : ', pi)
    print('\n')
```

# Internal Rate of Return

```python
def IRR() :
    print('INTERNAL RATE OF RETURN')
    ls = []
    n = int(input('Time Period : '))
    r = toFloat(input('Rate as (Rate/100) : '))
    sum_pv = 0
    ini = toFloat(input('Enter the inital investment : '))
    
    print('Enter year wise Cash Inflow : ')
    
    for yr in range(1,n+1):
        amount = toFloat(input())
        ls.append(amount)
   
    while(1):
        for yr in range(n):
            sum_pv = sum_pv + ls[yr] / (( 1 + r ) ** (yr+1))
        
        npv = sum_pv - ini

        if npv <= 0:
            break
        else:
            sum_pv = 0
            r = r + 0.0001

    print('Internal Rate of Return at NPV is 0 : {0} %' .format(r*100))
    print('\n')
```

# Payback Period

```python
def Payback():
    print('PAYBACK PERIOD WITHOUT DISCOUNTING')
    ls = []
    n = int(input('Time Period : '))
    r = toFloat(input('Rate as (Rate/100) : '))
    sum_pv = 0
    ini = toFloat(input('Inital investment : '))
    
    print('Enter year wise Cash Inflow : ')
    
    for yr in range(1,n+1):
        amount = toFloat(input())
        ls.append(amount)

    flag = 0
    for yr in range(n):
        pv = ls[yr] / (( 1 + r ) ** (yr+1))
        sum_pv = sum_pv + pv
        
        if sum_pv > ini:
            amt = sum_pv - pv
            period = yr + ((ini - amt) / pv)
            flag = 1
            break;
    
    if flag == 1 :
        print('Payback Period with Discounting : {0} years' .format(period))
    else :
        print('Investor will not be able to recover the complete amount that was invested')
    print('\n')
```
